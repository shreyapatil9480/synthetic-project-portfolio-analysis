What predicts satisfied enterprise clients?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/client_satisfaction.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
target = 'satisfied'
df[target].value_counts(normalize=True).plot(kind='bar', title='Target rate')
plt.tight_layout()
```

## Segment breakdown

```python
numeric = df.select_dtypes('number').columns
df[numeric].hist(figsize=(8,6))
plt.tight_layout()
```

## Modeling preview

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from src.features import prepare_features
X, y = prepare_features(df)
pipe = Pipeline([('s', StandardScaler()), ('m', LogisticRegression(max_iter=300))])
scores = cross_val_score(pipe, X, y, cv=3)
scores.mean()
```
